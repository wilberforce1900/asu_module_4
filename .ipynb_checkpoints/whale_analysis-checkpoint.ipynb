# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:

* `whale_returns.csv`: Contains returns of some famous "whale" investors' portfolios.

* `algo_returns.csv`: Contains returns from the in-house trading algorithms from Harold's company.

* `sp500_history.csv`: Contains historical closing prices of the S&P 500 Index.

## Whale Returns

Read the Whale Portfolio daily returns and clean the data

```python
# Reading whale returns
csv_whale_path = Path('Resources/whale_returns.csv')
df_whale = pd.read_csv(csv_whale_path)
```

```python
# Count nulls
## use isnull sum funcs
df_whale.isnull().sum()
```

```python
# Drop nulls
## use dropna func
df_whale.dropna()
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
# Reading algorithmic returns
csv_algo_path = Path('Resources/algo_returns.csv')
df_algo = pd.read_csv(csv_algo_path)
```

```python
# Count nulls
df_algo.isnull().sum()
```

```python
# Drop nulls
df_algo.dropna()
```

## S&P 500 Returns

Read the S&P 500 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P 500 Closing Prices
csv_sp500_path = Path('Resources/sp500_history.csv')
df_sp500 = pd.read_csv(csv_sp500_path)
```

```python
# Check Data Types
## use dtypes to check types
print(df_sp500)
print(df_sp500.dtypes)
```

```python
# Fix Data Types
df_sp500['Date'] = pd.to_datetime(df_sp500['Date'])
##float64 df['2017'] = df['2017'].apply(convert_currency), df['2016'].apply(lambda x: x.replace('$', '').replace(',', '')).astype('float')
## reference : https://pbpython.com/currency-cleanup.html
## use astype and also fix the $ from the str
df_sp500['Close'] = df_sp500['Close'].replace({'\$': '', ',': ''}, regex=True).astype(float)
print(df_sp500.dtypes)
```

```python
# Calculate Daily Returns
## use the function pct_change for daily returns
## reference: https://stackoverflow.com/questions/20000726/calculate-daily-returns-with-pandas-dataframe
daily_returns = df_sp500['Close'].pct_change(1)
#print(daily_returns)
```

```python
# Drop nulls
df_sp500.dropna()
```

```python
# Rename `Close` Column to be specific to this portfolio.
## use the rename function
df_sp500.rename(columns={'Close':'Closed Value'}, inplace = True)
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# Join Whale Returns, Algorithmic Returns, and the S&P 500 Returns into a single DataFrame with columns for each portfolio's returns.
## use the concat function
## reference: https://pandas.pydata.org/docs/user_guide/merging.html
pd_index = pd.concat([df_whale, df_algo, df_sp500], ignore_index = True, sort = False)

#pd_index = pd.concat([df_whale, df_algo, df_sp500], axis="columns", join="inner")
pd_index.head()
print(pd_index)
```

# Conduct Quantitative Analysis

In this section, you will calculate and visualize performance and risk metrics for the portfolios.

## Performance Anlysis

#### Calculate and Plot the daily returns.

```python
# Plot daily returns of all portfolios
daily_returns = pd_index['Closed Value'].pct_change(1)
daily_returns.plot()
```

#### Calculate and Plot cumulative returns.

```python
# Calculate cumulative returns of all portfolios
## use the compround function df.ix["Cumulative"] = ((df+1).cumprod()-1)
## reference: https://stackoverflow.com/questions/40811246/pandas-cumulative-return-function
cumu_returns = (pd_index['Closed Value']+1).cumprod()-1

# Plot cumulative returns
cumu_returns.plot()
```

```python
# Calculate  the daily standard deviation of S&P 500
## as daily_returns is the value of sp500, use std function for it
print(daily_returns.std())

# Determine which portfolios are riskier than the S&P 500
##use std to check every profolio
print(profolio.iloc[:,:1].std())
print(profolio.iloc[:,1:2].std())
print(profolio.iloc[:,2:3].std())
print(profolio.iloc[:,3:4].std())

# As the result shown below, the 'TIGER GLOBAL MANAGEMENT LLC', and 'BERKSHIRE HATHAWAY INC' come with higher risk than S&P500
```

```python
# Calculate the annualized standard deviation (252 trading days)
## locate the 252 trading days: as some data are NaN, choose the bottom 252 days
## notice that first get the values, and then use pct_change(1) to calculate the daily returns
## then use std to calculate
annu_sd = pd_index['Closed Value'].iloc[3697:3949].pct_change(1)
print(annu_sd.std())
```

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Calculate and plot the rolling standard deviation for all portfolios using a 21-day window
2. Calculate the correlation between each stock to determine which portfolios may mimick the S&P 500
3. Choose one portfolio, then calculate and plot the 60-day rolling beta between it and the S&P 500

### Calculate and plot rolling `std` for all portfolios with 21-day window

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window
## using the rolling to make the calculate
rolling_21 = daily_returns.rolling(window=21).std()
print(rolling_21)
# Plot the rolling standard deviation
rolling_21.plot()
```

```python
### Calculate and plot the correlation
import seaborn as sn
# Calculate the correlation
corr = pd_index.corr()
print(corr)
# Display de correlation matrix

sn.heatmap(corr, vmin = -1, vmax = 1)
```

### Calculate and Plot Beta for a chosen portfolio and the S&P 500

```python
# Calculate covariance of a single portfolio
## use the cov function
cov_1 = profolio.iloc[:,:1].cov()
cov_2 = profolio.iloc[:,1:2].cov()
cov_3 = profolio.iloc[:,2:3].cov()
cov_4 = profolio.iloc[:,3:4].cov()
print(cov_1)
print(cov_2)
print(cov_3)
print(cov_4)

# Calculate variance of S&P 500
## use var function
sp_var = daily_returns.var()
print(sp_var)

# Computing beta
## Beta = Covariance(Stock Returns, Market Returns) / Variance(Market Returns)
## source: https://stackoverflow.com/questions/59110902/is-there-anyway-to-calculate-market-beta-from-yahoo-finance-datasreader-on-pytho
beta = (profolio.cov())/(sp_var)

# Plot beta trend
beta.plot()
```
