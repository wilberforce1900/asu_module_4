## S&P 500 Returns

Read the S&P 500 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P 500 Closing Prices
csv_sp500_path = Path('Resources/sp500_history.csv')
df_sp500 = pd.read_csv(csv_sp500_path, index_col = "Date", infer_datetime_format = True, parse_dates = True)
```

```python
# Check Data Types
## use dtypes to check types
print(df_sp500)
print(df_sp500.dtypes)
```

```python
# Fix Data Types
#df_sp500['Date'] = pd.to_datetime(df_sp500['Date']), no need to fix date type since other df are same
#df_sp500
##float64 df['2017'] = df['2017'].apply(convert_currency), df['2016'].apply(lambda x: x.replace('$', '').replace(',', '')).astype('float')
## reference : https://pbpython.com/currency-cleanup.html
## use astype and also fix the $ from the str
df_sp500['Close'] = df_sp500['Close'].replace({'\$': '', ',': ''}, regex=True).astype(float)
df_sp500
print(df_sp500)
print(df_sp500.dtypes)
```

```python
# Calculate Daily Returns
## use the function pct_change for daily returns
## reference: https://stackoverflow.com/questions/20000726/calculate-daily-returns-with-pandas-dataframe
df_sp500['Close'] = df_sp500['Close'].pct_change()
df_sp500
```

```python
# Drop nulls
df_sp500 = df_sp500.dropna()
df_sp500
```

```python
# Rename `Close` Column to be specific to this portfolio.
## use the rename function pd.rename(columns = {'team':'team_name', 'points':'points_scored'}, inplace = True)
df_sp500.columns = ['S&P TSX']
df_sp500
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# Join Whale Returns, Algorithmic Returns, and the S&P 500 Returns into a single DataFrame with columns for each portfolio's returns.
## use the concat function
## reference: https://pandas.pydata.org/docs/user_guide/merging.html
#pd_index = pd.concat([df_whale, df_algo, df_sp500], ignore_index = True, sort = False)
#pd.concat([df_whale, df_algo, df_sp500], axis="columns", join="inner")

pd_index = pd.concat([df_whale, df_algo, df_sp500],axis="columns", join="inner")
#pd_index = pd_index.head()
pd_index.head()
```

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Calculate and plot the rolling standard deviation for all portfolios using a 21-day window
2. Calculate the correlation between each stock to determine which portfolios may mimick the S&P 500
3. Choose one portfolio, then calculate and plot the 60-day rolling beta between it and the S&P 500

### Calculate and plot rolling `std` for all portfolios with 21-day window

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window
## using the rolling to make the calculate
rolling_21 = pd_index.rolling(window=21).std()
rolling_21
#print(rolling_21)
# Plot the rolling standard deviation
rolling_21.plot()
```

```python
### Calculate and plot the correlation
import seaborn as sn
# Calculate the correlation
corr = pd_index.corr()
corr
# Display de correlation matrix

sn.heatmap(corr, vmin = -1, vmax = 1)
```

### Calculate and Plot Beta for a chosen portfolio and the S&P 500

```python
# Calculate covariance of a single portfolio
## use the cov function

cov_val = pd_index.cov()
cov_val
print(cov_val)

# Calculate variance of S&P 500
## use var function
sp_var = pd_index.var()
sp_var
print(sp_var)

# Computing beta
## Beta = Covariance(Stock Returns, Market Returns) / Variance(Market Returns)
## source: https://stackoverflow.com/questions/59110902/is-there-anyway-to-calculate-market-beta-from-yahoo-finance-datasreader-on-pytho
beta = (cov_val)/(sp_var)
beta
# Plot beta trend
beta.plot()
```

## Rolling Statistics Challenge: Exponentially Weighted Average 

An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the [`ewm`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.ewm.html) with a 21-day half life for each portfolio, using standard deviation (`std`) as the metric of interest.

```python
# Use `ewm` to calculate the rolling window
## source: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.ewm.html

pd_index.ewm(halflife = 21).std()
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios
## https://www.learnpythonwithrune.org/how-to-calculate-sharpe-ratio-with-pandas-and-numpy/
##sharpe_ratio = log_return.mean()/log_return.std()


shapre_annu = (pd_index.pct_change().mean() * 252 )/(pd_index.pct_change().std() * np.sqrt(252))
shapre_annu
```

```python
# Visualize the sharpe ratios as a bar plot
shapre_annu.plot.bar()
```
